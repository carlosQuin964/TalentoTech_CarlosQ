```python
import requests
import pandas as pd
import plotly.graph_objects as go

# Lista de ciudades con su respectivo país
ciudades = [
    ('Bogotá', 'Colombia'),
    ('Buenos Aires', 'Argentina'),
    ('Lima', 'Perú'),
    ('Manaus', 'Brasil'),
    ('Santiago de Chile', 'Chile'),
    ('Sao Paulo', 'Brasil'),
    ('Ushuaia', 'Argentina'),
    ('Recife', 'Brasil'),
    ('La Habana', 'Cuba'),
    ('New York', 'United States'),
    ('Los Angeles', 'United States'),
    ('Seattle', 'United States'),
]

# Lista para almacenar los datos de temperatura
datos_temperatura = []

for ciudad, pais in ciudades:
    url = f'http://api.openweathermap.org/data/2.5/weather?q={ciudad},{pais}&appid={API_KEY}&units=metric'
    respuesta = requests.get(url)

    if respuesta.status_code == 200:
        datos = respuesta.json()
        if 'main' in datos:
            temperatura_actual = datos['main']['temp']
            latitud = datos['coord']['lat']
            longitud = datos['coord']['lon']
            datos_temperatura.append({'Ciudad': ciudad, 'Pais': pais, 'Temperatura': temperatura_actual, 'Latitud': latitud, 'Longitud': longitud})
        else:
            print(f'Error: No se encontraron los datos de temperatura para {ciudad}, {pais}.')
    else:
        print(f'Error: No se pudo obtener los datos del clima para {ciudad}, {pais}. Código de estado: {respuesta.status_code}')

# Crear un DataFrame de Pandas con los datos de temperatura
df_temperatura = pd.DataFrame(datos_temperatura)

# Crear un mapa usando Plotly
fig = go.Figure()

# Añadir puntos al mapa con los textos formateados
fig.add_trace(go.Scattermapbox(
    lat=df_temperatura['Latitud'],
    lon=df_temperatura['Longitud'],
    mode='markers+text',
    marker=go.scattermapbox.Marker(
        size=14,
        color='blue'
    ),
    text=df_temperatura['Ciudad'],  # Texto con el nombre de la ciudad
    texttemplate='%{text} <br>%{customdata}°C',  # Mostrar ciudad y temperatura
    customdata=df_temperatura['Temperatura'],  # Usar la temperatura como datos personalizados
    textfont=dict(size=14, color="black"),
    textposition='top right'
))

# Configuración del mapa con CartoDB Positron
fig.update_layout(
    mapbox=dict(
        style='carto-positron',  # CartoDB Positron style for a clean background
        zoom=1.5,
        center=dict(lat=-4.5, lon=-58.5),  # Centro de Sudamérica
    ),
    margin={"r":0,"t":0,"l":0,"b":0}
)

# Mostrar el mapa
fig.show()
```

```python
import requests
import pandas as pd
from datetime import datetime

# Lista de ciudades para obtener datos del clima
ciudades = [
    ('Bogotá', 'Colombia'),
    ('Buenos Aires', 'Argentina'),
    ('Lima', 'Perú'),
    ('Manaus', 'Brasil'),
    ('Santiago de Chile', 'Chile'),
    ('Sao Paulo', 'Brasil'),
    ('Ushuaia', 'Argentina'),
    ('Recife', 'Brasil'),
    ('La Habana', 'Cuba'),
    ('New York', 'United States'),
    ('Los Angeles', 'United States'),
    ('Seattle', 'United States'),
]

# Lista para almacenar los datos actuales del clima
datos_clima_actual = []

# Recorrer cada ciudad y obtener los datos del clima actual
for ciudad, pais in ciudades:
    url = f'http://api.openweathermap.org/data/2.5/weather?q={ciudad},{pais}&appid={API_KEY}&units=metric'

    try:
        respuesta = requests.get(url)
        respuesta.raise_for_status()  # Lanza una excepción para errores HTTP

        if respuesta.status_code == 200:
            datos = respuesta.json()

            # Extracción de los datos
            temperatura = datos['main']['temp']
            humedad = datos['main']['humidity']
            presion = datos['main']['pressure']
            velocidad_viento = datos['wind']['speed']
            visibilidad = datos['visibility'] / 1000  # Convertir de metros a kilómetros
            nubes = datos['clouds']['all']
            descripcion_clima = datos['weather'][0]['description']
            salida_sol = datetime.utcfromtimestamp(datos['sys']['sunrise']).strftime('%H:%M:%S')
            puesta_sol = datetime.utcfromtimestamp(datos['sys']['sunset']).strftime('%H:%M:%S')

            # Agregar los datos de la ciudad a la lista
            datos_clima_actual.append({
                'Ciudad': ciudad,
                'País': pais,
                'Temperatura (°C)': temperatura,
                'Humedad (%)': humedad,
                'Presión (hPa)': presion,
                'Velocidad del Viento (m/s)': velocidad_viento,
                'Visibilidad (km)': visibilidad,
                'Nubes (%)': nubes,
                'Clima': descripcion_clima,
                'Amanecer': salida_sol,
                'Atardecer': puesta_sol
            })
        else:
            print(f'Error: No se pudo obtener datos para {ciudad}. Código de estado: {respuesta.status_code}')

    except requests.exceptions.RequestException as e:
        print(f"Error de solicitud para {ciudad}: {e}")

# Crear un DataFrame para mostrar los datos en forma de tabla
df_clima_actual = pd.DataFrame(datos_clima_actual)

# Mostrar la tabla (datos del clima actual)
df_clima_actual.style.set_table_styles(
    [{'selector': 'thead th', 'props': [('background-color', '#f2f2f2'), ('color', 'black'), ('font-weight', 'bold')]},
     {'selector': 'tbody td', 'props': [('background-color', '#fff'), ('color', 'black')]}]
)

# Mostrar la tabla directamente en el notebook
df_clima_actual
```
